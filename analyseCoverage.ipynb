```python
import csv
import os

for project in t_evos_buggy_commits.keys():
    for buggy_commit in t_evos_buggy_commits[project].keys():
        buggy_files = t_evos_buggy_commits[project][buggy_commit]
        csv_file_path = coverage_folder + project+ "/" + buggy_commit +'.csv'
        if os.path.exists(csv_file_path):
            coverages = {}
            with open(os.path.join(csv_file_path), 'r') as file:
                csv_reader = csv.reader(file)
                count = 0
                for row in csv_reader:
                    if count > 0:
                        coverages[row[0]] = row[1]
                    count = count + 1
            file.close()
            values = coverages.values()
            values = [float(x) for x in values]
            coverages_rank = sorted(values)
            print(project + " - " + buggy_commit)
            for file in buggy_files:
                try:
                    file_coverage = coverages[file]
                    print(file + " ----> coverage = " + str(file_coverage))
                    print(file + " ----> rank - " + str(coverages_rank.index(float(file_coverage)) + 2) + " out of " + str(len(coverages_rank) + 1))

                except: # No coverage at all
                    print(file + " ----> coverage = 0")
                    print(file + " ----> rank - 1 out of " + str(len(coverages_rank) + 1))
            print("---")
        else:
            print("Skipping commit " + buggy_commit + " from project " + project + " - CSV file not found")
            print("----")
```
