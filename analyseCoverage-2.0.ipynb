```python
import os
import glob
import json

def get_file_line_from_stack_trace(stack_trace_entry):
    try: # Treating exceptions such as Unknow Source and Native Method
        colon_index = stack_trace_entry.index(':')
        parentheses_closing_index = stack_trace_entry.index(')', colon_index + 1)
        file_line = stack_trace_entry[colon_index + 1 : parentheses_closing_index]
        return file_line
    except:
        return -1


def get_file_name_from_stack_trace(stack_trace_entry):
    try: #  Treating exceptions such as Unknow Source and Native Method
        parentheses_opening_index = stack_trace_entry.index('(')
        colon_index = stack_trace_entry.index(':',parentheses_opening_index + 1)
        file_name = stack_trace_entry[parentheses_opening_index + 1 : colon_index]
    except:
        parentheses_opening_index = stack_trace_entry.index('(')
        parentheses_closing_index = stack_trace_entry.index(')',parentheses_opening_index + 1)
        file_name = stack_trace_entry[parentheses_opening_index + 1 : parentheses_closing_index]
    if ("." in file_name):
        file_name = file_name.split(".")[0]
    return file_name

def get_method_name_from_stack_trace(stack_trace_entry):
    parentheses_opening_index = stack_trace_entry.index('(')
    stack_trace_entry = stack_trace_entry[:parentheses_opening_index]
    last_dot_index = stack_trace_entry.rfind(".")
    method_name = stack_trace_entry[last_dot_index + 1 : ]
    return method_name

def json_file_to_dict(file):
    data = {}
    with open(os.path.join(file), 'r') as fp:
        data = json.load(fp)
    fp.close()
    return data

def find_file_covered_lines(test_data, file_name):
    for test_file in test_data["covered_lines"].keys():
        if test_file.endswith(file_name):
            return test_data["covered_lines"][test_file]
    return []

def dict_to_json_file(file, dic, folder):
    if not os.path.exists(folder):
        os.makedirs(folder)
    with open(os.path.join(folder, file+'.json'), 'w') as fp:
        json.dump(dic, fp, sort_keys=True, indent=4)
    fp.close()

def check_if_method_is_covered_by_test(test_data, file_name, method_name):
    for key in test_data["covered_stats"].keys():
        if(method_name + "(") in key:
            return True
    return False


def get_files_that_cover_the_method(file_name, method_name, coverage_details_folder):
    tests_list = []
    for test_coverage_file in glob.glob(coverage_details_folder + "*.json"):
        test_file_name = test_coverage_file.split("/")[-1]
        if (test_file_name != "coverage_details.json"):
            test_data = json_file_to_dict(test_coverage_file)
            if (check_if_method_is_covered_by_test(test_data, file_name, method_name)):
                tests_list.append(test_file_name)
    return tests_list

def run_checkout(project_path, buggy_commit):
    os.chdir(project_path)
    checkout_command = "git checkout  --quiet " + buggy_commit
    os.system(checkout_command)

def find_file(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)

def read_file_lines(file_name, project_path):
    path = find_file(file_name, project_path)
    if path:
        with open(path, 'r', encoding='utf-8', errors='ignore') as file:
            lines =file.readlines()
        file.close()
        return lines
    return []

def line_contain_method_declaration(pos_line, lines, method_name):
    line = lines[pos_line]
    if not (" " + method_name + "(" in line or " " + method_name + " (" in line):
        return False

    if " = " in line:
        return False

    found = False

    for i in range(pos_line, len(lines)):
        if "{" in lines[i]:
            if "}" in lines[i]:
                if lines[i].find("}") < lines[i].find("{"):
                    return False
                else:
                    found = True
                    break
            else:
                found = True
                break
        else:
            if "}" in lines[i]:
                return False

    if not found:
        return False

    return True

def closest (num, arr):
    curr = arr[0]
    for element in arr:
        if abs (num - int(element)) < abs (num - curr):
            curr = element
    return curr

def find_closing_braces (lines, pos_init_line):
    count = 0
    pos = pos_init_line
    found_opening_braces_already = False
    while (pos < len(lines)):
        if not found_opening_braces_already:
            if "{" in lines[pos]:
                found_opening_braces_already = True
        count += lines[pos].count("{")
        count -= lines[pos].count("}")
        if count == 0 and found_opening_braces_already:
            break
        pos += 1

    return min(pos, len(lines) - 1)

def get_method_begin_and_end_lines(file_name, method_name,  project, project_path, buggy_commit, file_line):
    run_checkout(project_path, buggy_commit)
    lines = read_file_lines(file_name+".java", project_path)
    possible_init_lines = []
    for i in range(0, len(lines)):
        if line_contain_method_declaration(i, lines, method_name):
            possible_init_lines.append(i + 1) # Do not starts with 0 but 1

    if possible_init_lines == []:
        return(-1, -1)

    init_line = closest(int(file_line), possible_init_lines)

    final_line = find_closing_braces (lines, init_line - 1) + 1
    return(init_line, final_line)

def get_method_covered_lines(file_name, begin_line, end_line, coverage_details_folder):
    file_covered_lines = []
    for test_coverage_file in glob.glob(coverage_details_folder + "*.json"):
        test_name = os.path.basename(test_coverage_file).replace('.json', '')
        if(test_name == "coverage_details"):
            continue
        test_data = json_file_to_dict(test_coverage_file)
        current_test_file_covered_lines = find_file_covered_lines(test_data, file_name)
        file_covered_lines.extend(current_test_file_covered_lines)

    file_covered_lines = sorted(list(set(file_covered_lines)))

    method_covered_lines = []
    for l in file_covered_lines:
        if (int(l) >= begin_line and int(l) <= end_line):
            method_covered_lines.append(int(l))
    return method_covered_lines

def count_methods_lines_of_code(begin_line, end_line, file_name, project_path, method_covered_lines):

    lines = read_file_lines(file_name+".java", project_path)
    count = 0
    multiline_comment_active = False
    begin_pos = begin_line - 1
    end_pos = end_line - 1

    for i in range(begin_pos + 1, end_pos + 1):
        line = lines[i]

        # If it is covered, it is a code line
        if i + 1 in method_covered_lines:
            count +=1
            continue

        # Ignore blocks inside multiline comments
        if line.strip().startswith("/*"):
            multiline_comment_active = True
            continue
        if multiline_comment_active:
            if line.strip().endswith("*/"):
                multiline_comment_active = False
            continue
        # Ignore single line comments
        if line.strip().startswith("//"):
            continue

        # Remove end-ine comments
        line = line.split("//")[0]
        # Count non-empty lines
        if line.strip() != "":
            # Ignoring closing braces only lines
            if line.strip() == "}":
                continue

            # Ignoring conditional lines (branches)
            if line.strip().startswith("if ") or line.strip().startswith("else ") or line.strip().startswith("else:"):
                continue

            # Ignoring loop definition lines
            if line.strip().startswith("for ") or line.strip().startswith("while "):
                continue

            # Treating statements with line break
            if line.strip().endswith("{") or line.strip().endswith("}") or line.strip().endswith(";") or line.strip().endswith(","):
                count = count + 1

    return count

def get_method_coverage(file_name, method_name, file_line, coverage_details_folder, project, buggy_commit):

    project_path = base_path + "open_source_repos_being_studied/"+  project + "/"
    (begin_line, end_line) = get_method_begin_and_end_lines(file_name, method_name,  project, project_path, buggy_commit, file_line)

    if begin_line == -1 and end_line== -1:
        return 0.0

    method_covered_lines = get_method_covered_lines(file_name, begin_line, end_line, coverage_details_folder)
    method_len = count_methods_lines_of_code(begin_line, end_line, file_name, project_path, method_covered_lines)

    method_coverage = (len(method_covered_lines) / method_len) * 100
    return method_coverage

def check_if_file_is_in_source_code(file_name, project, buggy_commit):
    project_path = base_path + "open_source_repos_being_studied/"+  project + "/"
    run_checkout(project_path, buggy_commit)
    path = find_file(file_name + ".java", project_path)
    if path:
        return True
    return False

def add_unique_tests_that_cover_the_methods (bug_report_data):
    st_entries = list(bug_report_data.keys())
    source_code_st_entries = []
    for entry in st_entries:
        if  bug_report_data[entry]["file_was_found_in_source_code"]:
            source_code_st_entries.append(entry)

    repeated_test_methods = []
    if len(source_code_st_entries) > 1:
        first_st_entry = source_code_st_entries[0]
        repeated_test_methods = list(bug_report_data[first_st_entry]["tests_that_cover_the_method"])

        for i in range(1, len(source_code_st_entries)):
            current_stack_trace = source_code_st_entries[i]
            removal_list = []
            for test in repeated_test_methods:
                if test not in bug_report_data[current_stack_trace]["tests_that_cover_the_method"]:
                    removal_list.append(test)
            for test in removal_list:
                repeated_test_methods.remove(test)

    result = {}
    for current_st_entry in source_code_st_entries:
        unique_test_methods = list(bug_report_data[current_st_entry]["tests_that_cover_the_method"])
        removal_list = []
        for test in unique_test_methods:
            if test in repeated_test_methods:
                removal_list.append(test)
        for test in removal_list:
            unique_test_methods.remove(test)
        result[current_st_entry] = unique_test_methods
    return result

for project in t_evos_bugs_information.keys():
    print("!!!!")
    print(project)

    for bug_report_id in t_evos_bugs_information[project].keys():

        buggy_commit = t_evos_bugs_information[project][bug_report_id]["commit_with_bug"]
        buggy_files = t_evos_bugs_information[project][bug_report_id]["buggy_files"]
        stack_trace_entries = t_evos_bugs_information[project][bug_report_id]["stack_trace_entries"]
        coverage_details_folder = coverage_data_folder + project + "/" + buggy_commit + "/details/"

        print("----")
        print(bug_report_id)
        print(buggy_commit)


        if (not os.path.exists(coverage_details_folder)):
            print("Skipping bug report " + bug_report_id + " from project " + project + " - CSV file not found")
            print("----")
            continue
        stack_trace_entries_structured = []
        stack_trace_analysis = {}
        for st_entry in stack_trace_entries:
            file_line = get_file_line_from_stack_trace(st_entry)
            file_name = get_file_name_from_stack_trace(st_entry)
            method_name = get_method_name_from_stack_trace(st_entry)
            stack_trace_entries_structured.append(
                {
                    "stack_trace": st_entry,
                    "file_line": file_line,
                    "file_name": file_name,
                    "method_name": method_name
                }
            )
            file_is_in_source_code = check_if_file_is_in_source_code(file_name, project, buggy_commit)
            list_of_tests_covering_method = get_files_that_cover_the_method(file_name, method_name, coverage_details_folder) if file_is_in_source_code else "N/A"
            number_of_tests_covering_method = len(list_of_tests_covering_method) if file_is_in_source_code else "N/A"
            stack_trace_analysis[st_entry] = {
                "tests_that_cover_the_method": list_of_tests_covering_method,
                "number_of_tests_that_cover_the_method": number_of_tests_covering_method,
                "tests_that_cover_the_statement": [],
                "number_of_tests_that_cover_the_statement": 0,
                "method_coverage": get_method_coverage(file_name, method_name, file_line, coverage_details_folder, project, buggy_commit)  if file_is_in_source_code else "N/A",
                "method_is_buggy": file_name + "." + method_name in t_evos_bugs_information[project][bug_report_id]["buggy_methods"],
                "file_is_buggy": (file_name + ".java") in t_evos_bugs_information[project][bug_report_id]["buggy_files"],
                "file_was_found_in_source_code": file_is_in_source_code
            }
        for test_coverage_file in glob.glob(coverage_details_folder + "*.json"):
            test_name = os.path.basename(test_coverage_file).replace('.json', '')
            if(test_name == "coverage_details"):
                continue
            test_data = json_file_to_dict(test_coverage_file)
            for st in stack_trace_entries_structured:
                stack_trace = st["stack_trace"]
                file_found_in_source_code = stack_trace_analysis[stack_trace]["file_was_found_in_source_code"]
                if not file_found_in_source_code:
                    stack_trace_analysis[stack_trace]["tests_that_cover_the_statement"] = "N/A"
                    stack_trace_analysis[stack_trace]["number_of_tests_that_cover_the_statement"] = "N/A"
                else:
                    if st["file_line"] != -1:
                        file_covered_lines = find_file_covered_lines(test_data, st["file_name"])
                        if st["file_line"] in file_covered_lines:
                            stack_trace_analysis[stack_trace]["tests_that_cover_the_statement"].append(test_name)
                            stack_trace_analysis[stack_trace]["number_of_tests_that_cover_the_statement"] += 1
        if bug_report_id == "CSV-100":
            print("HI")
        unique_tests_object = add_unique_tests_that_cover_the_methods(stack_trace_analysis)
        for st_entry in unique_tests_object.keys():
            stack_trace_analysis[st_entry]["unique_tests_that_cover_the_method"] = unique_tests_object[st_entry]
        dict_to_json_file(bug_report_id, stack_trace_analysis, out + project + "/")

print("Extraction completed")
```
