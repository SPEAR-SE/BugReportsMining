```python
import glob
import os
import json
import csv


def json_file_to_dict(file):
    data = {}
    with open(os.path.join(file), 'r') as fp:
        data = json.load(fp)
    fp.close()
    return data

def dict_to_json_file(file, dic, folder):
    if not os.path.exists(folder):
        os.makedirs(folder)
    with open(os.path.join(folder, file+'.json'), 'w') as fp:
        json.dump(dic, fp, sort_keys=True, indent=4)
    fp.close()


def sort_dict (dictionary):
    return dict(sorted(dictionary.items(), key=lambda item: item[1], reverse = True))

def get_best_position_buggy_method (ranking_data, buggy_methods_list):
    best_position = len(ranking_data) + 1
    for method in buggy_methods_list:
        try:
            if ranking_data[method] < best_position:
                best_position = ranking_data[method]
        except:
            continue
    return best_position

def run_checkout(project_path, buggy_commit):
    os.chdir(project_path)
    checkout_command = "git checkout  --quiet " + buggy_commit
    os.system(checkout_command)

def find_file(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)

def read_file_lines(file_name, project_path):
    path = find_file(file_name, project_path)
    if path:
        with open(path, 'r', encoding='utf-8', errors='ignore') as file:
            lines =file.readlines()
        file.close()
        return lines
    return []

def line_contain_method_declaration(pos_line, lines, method_name):
    line = lines[pos_line]
    if not (" " + method_name + "(" in line or " " + method_name + " (" in line):
        return False

    if " = " in line:
        return False

    found = False

    for i in range(pos_line, len(lines)):
        if "{" in lines[i]:
            if "}" in lines[i]:
                if lines[i].find("}") < lines[i].find("{"):
                    return False
                else:
                    found = True
                    break
            else:
                found = True
                break
        else:
            if "}" in lines[i]:
                return False

    if not found:
        return False

    return True

def closest (num, arr):
    curr = arr[0]
    for element in arr:
        if abs (num - int(element)) < abs (num - curr):
            curr = element
    return curr

def find_closing_braces (lines, pos_init_line):
    count = 0
    pos = pos_init_line
    found_opening_braces_already = False
    while (pos < len(lines)):
        if not found_opening_braces_already:
            if "{" in lines[pos]:
                found_opening_braces_already = True
        count += lines[pos].count("{")
        count -= lines[pos].count("}")
        if count == 0 and found_opening_braces_already:
            break
        pos += 1

    return min(pos, len(lines) - 1)

def get_method_begin_and_end_lines(file_name, method_name,  project, project_path, buggy_commit, file_line = None):
    run_checkout(project_path, buggy_commit)
    lines = read_file_lines(file_name+".java", project_path)
    possible_init_lines = []
    for i in range(0, len(lines)):
        if line_contain_method_declaration(i, lines, method_name):
            possible_init_lines.append(i + 1) # Do not starts with 0 but 1

    if possible_init_lines == []:
        return(-1, -1)

    if file_line:
        init_line = closest(int(file_line), possible_init_lines)
    else:
        init_line = possible_init_lines[0]

    final_line = find_closing_braces (lines, init_line - 1) + 1
    return(init_line, final_line)

def get_method_covered_lines(file_name, begin_line, end_line, coverage_details_folder):
    file_covered_lines = []
    for test_coverage_file in glob.glob(coverage_details_folder + "*.json"):
        test_name = os.path.basename(test_coverage_file).replace('.json', '')
        if(test_name == "coverage_details"):
            continue
        test_data = json_file_to_dict(test_coverage_file)
        current_test_file_covered_lines = find_file_covered_lines(test_data, file_name)
        file_covered_lines.extend(current_test_file_covered_lines)

    file_covered_lines = sorted(list(set(file_covered_lines)))

    method_covered_lines = []
    for l in file_covered_lines:
        if (int(l) >= begin_line and int(l) <= end_line):
            method_covered_lines.append(int(l))
    return method_covered_lines


def find_file_covered_lines(test_data, file_name):
    for test_file in test_data["covered_lines"].keys():
        if test_file.endswith(file_name):
            return test_data["covered_lines"][test_file]
    return []

def count_methods_lines_of_code(begin_line, end_line, file_name, project_path, method_covered_lines):

    lines = read_file_lines(file_name+".java", project_path)
    count = 0
    multiline_comment_active = False
    begin_pos = begin_line - 1
    end_pos = end_line - 1

    for i in range(begin_pos, end_pos + 1):
        line = lines[i]

        # If it is covered, it is a code line
        if i + 1 in method_covered_lines:
            count +=1
            continue

        # Ignore blocks inside multiline comments
        if line.strip().startswith("/*"):
            multiline_comment_active = True
            continue
        if multiline_comment_active:
            if line.strip().endswith("*/"):
                multiline_comment_active = False
            continue
        # Ignore single line comments
        if line.strip().startswith("//"):
            continue

        # Remove end-ine comments
        line = line.split("//")[0]
        # Count non-empty lines
        if line.strip() != "":
            # Ignoring closing braces only lines
            if line.strip() == "}":
                continue

            # Ignoring conditional lines (branches)
            if line.strip().startswith("if ") or line.strip().startswith("else ") or line.strip().startswith("else:"):
                continue

            # Ignoring loop definition lines
            if line.strip().startswith("for ") or line.strip().startswith("while "):
                continue

            # Treating statements with line break
            if line.strip().endswith("{") or line.strip().endswith("}") or line.strip().endswith(";") or line.strip().endswith(","):
                count = count + 1

    return count

def check_if_file_is_in_source_code(file_name, project, buggy_commit):
    project_path = base_path + "open_source_repos_being_studied/"+  project + "/"
    run_checkout(project_path, buggy_commit)
    path = find_file(file_name + ".java", project_path)
    if path:
        return True
    return False

def get_method_coverage(file_name, method_name, file_line, coverage_details_folder, project, buggy_commit):

    project_path = base_path + "open_source_repos_being_studied/"+  project + "/"
    (begin_line, end_line) = get_method_begin_and_end_lines(file_name, method_name,  project, project_path, buggy_commit, file_line)

    if begin_line == -1 and end_line== -1:
        return 0.0

    method_covered_lines = get_method_covered_lines(file_name, begin_line, end_line, coverage_details_folder)
    method_len = count_methods_lines_of_code(begin_line, end_line, file_name, project_path, method_covered_lines)

    if method_len == 0:
        print("method with 0 lines")
        return 0

    method_coverage = (len(method_covered_lines) / method_len) * 100

    return method_coverage


def get_method_name_from_stack_trace(stack_trace_entry):
    parentheses_opening_index = stack_trace_entry.index('(')
    stack_trace_entry = stack_trace_entry[:parentheses_opening_index]
    last_dot_index = stack_trace_entry.rfind(".")
    method_name = stack_trace_entry[last_dot_index + 1 : ]
    return method_name

def get_file_line_from_stack_trace(stack_trace_entry):
    try: # Treating exceptions such as Unknow Source and Native Method
        colon_index = stack_trace_entry.index(':')
        parentheses_closing_index = stack_trace_entry.index(')', colon_index + 1)
        file_line = stack_trace_entry[colon_index + 1 : parentheses_closing_index]
        return file_line
    except:
        return -1

def get_file_name_from_stack_trace(stack_trace_entry):
    try: #  Treating exceptions such as Unknow Source and Native Method
        parentheses_opening_index = stack_trace_entry.index('(')
        colon_index = stack_trace_entry.index(':',parentheses_opening_index + 1)
        file_name = stack_trace_entry[parentheses_opening_index + 1 : colon_index]
    except:
        parentheses_opening_index = stack_trace_entry.index('(')
        parentheses_closing_index = stack_trace_entry.index(')',parentheses_opening_index + 1)
        file_name = stack_trace_entry[parentheses_opening_index + 1 : parentheses_closing_index]
    if ("." in file_name):
        file_name = file_name.split(".")[0]
    return file_name

def print_best_buggy_method_positions_for_each_Ochiai():
    for ochiai_output_folder in ochiai_scores_paths_list:
        for bug_report_analysis_file in glob.glob(ochiai_output_folder + "*/*.json"):
            project = bug_report_analysis_file.split("/")[-2]
            bug_id = bug_report_analysis_file.split("/")[-1].replace(".json","")
            ochiai_identificator = bug_report_analysis_file.split("/")[-3]
            if project not in t_evos_bugs_information.keys():
                continue
            if bug_id not in t_evos_bugs_information[project].keys():
                continue
            print("---------------")
            print(ochiai_identificator + " - " + project + " - " + bug_id)
            ochiai_scores_data = json_file_to_dict(bug_report_analysis_file)
            ochiai_scores_data = sort_dict(ochiai_scores_data)

            ranking = {}
            actual_pos = 1
            repetition_count = 0
            previous_score = -1
            for test in ochiai_scores_data.keys():
                if ochiai_scores_data[test] != previous_score:
                    actual_pos = actual_pos + repetition_count
                    repetition_count = 0
                    previous_score = ochiai_scores_data[test]
                    ranking [test] = actual_pos
                    actual_pos += 1
                else:
                    repetition_count +=1
                    ranking [test] = actual_pos

            first_buggy_method_position = get_best_position_buggy_method (ranking, t_evos_bugs_information[project][bug_id]["buggy_methods"])
            print(first_buggy_method_position)

def print_average_coverage_of_the_methods_in_the_stack_trace():
    for project in t_evos_bugs_information.keys():
        for bug_id in t_evos_bugs_information[project].keys():
            print("---------------")
            print(project + " - " + bug_id)
            buggy_commit = t_evos_bugs_information[project][bug_id]["commit_with_bug"]
            coverage_details_folder = coverage_data_folder + project + "/" + buggy_commit + "/details/"
            coverage_sum = 0
            internal_entries_number = 0
            stack_trace_entries = t_evos_bugs_information[project][bug_id]["stack_trace_entries"]
            for st_entry in stack_trace_entries:
                file_name = get_file_name_from_stack_trace(st_entry)
                method_name = get_method_name_from_stack_trace(st_entry)
                file_line = get_file_line_from_stack_trace(st_entry)
                if check_if_file_is_in_source_code(file_name, project, buggy_commit):
                    coverage_sum += get_method_coverage(file_name, method_name, file_line, coverage_details_folder, project, buggy_commit)
                    internal_entries_number += 1
            if internal_entries_number == 0:
                print ("N/A")
            else:
                coverage_average = coverage_sum/internal_entries_number
                print(coverage_average)

def print_average_coverage_of_the_files_in_the_stack_trace():
    for project in t_evos_bugs_information.keys():
        for bug_id in t_evos_bugs_information[project].keys():
            print("---------------")
            print(project + " - " + bug_id)
            buggy_commit = t_evos_bugs_information[project][bug_id]["commit_with_bug"]
            csv_file_path = coverage_folder + project+ "/" + buggy_commit +'.csv'
            if os.path.exists(csv_file_path):
                coverages = {}
                with open(os.path.join(csv_file_path), 'r') as file:
                    csv_reader = csv.reader(file)
                    count = 0
                    for row in csv_reader:
                        if count > 0:
                            coverages[row[0]] = row[1]
                        count = count + 1
            coverage_sum = 0
            internal_entries_number = 0
            stack_trace_entries = t_evos_bugs_information[project][bug_id]["stack_trace_entries"]
            for st_entry in stack_trace_entries:
                file_name = get_file_name_from_stack_trace(st_entry)
                if check_if_file_is_in_source_code(file_name, project, buggy_commit):
                    coverage_sum += float(coverages[file_name + ".java"])
                    internal_entries_number += 1
            if internal_entries_number == 0:
                print ("N/A")
            else:
                coverage_average = coverage_sum/internal_entries_number
                print(coverage_average)

def print_average_coverage_of_the_buggy_methods():
    for project in t_evos_bugs_information.keys():
        for bug_id in t_evos_bugs_information[project].keys():
            print("---------------")
            print(project + " - " + bug_id)
            buggy_commit = t_evos_bugs_information[project][bug_id]["commit_with_bug"]
            coverage_details_folder = coverage_data_folder + project + "/" + buggy_commit + "/details/"
            coverage_sum = 0
            buggy_methods = t_evos_bugs_information[project][bug_id]["buggy_methods"]
            for bm in buggy_methods:
                file_name = bm.split(".")[0]
                method_name = bm.split(".")[1]
                coverage_sum += get_method_coverage(file_name, method_name, None, coverage_details_folder, project, buggy_commit)

            coverage_average = coverage_sum/len(buggy_methods)
            print(coverage_average)

def print_average_coverage_of_the_buggy_files():
    for project in t_evos_bugs_information.keys():
        for bug_id in t_evos_bugs_information[project].keys():
            print("---------------")
            print(project + " - " + bug_id)
            buggy_commit = t_evos_bugs_information[project][bug_id]["commit_with_bug"]
            csv_file_path = coverage_folder + project+ "/" + buggy_commit +'.csv'
            if os.path.exists(csv_file_path):
                coverages = {}
                with open(os.path.join(csv_file_path), 'r') as file:
                    csv_reader = csv.reader(file)
                    count = 0
                    for row in csv_reader:
                        if count > 0:
                            coverages[row[0]] = row[1]
                        count = count + 1
            coverage_sum = 0
            buggy_methods = t_evos_bugs_information[project][bug_id]["buggy_methods"]
            for bm in buggy_methods:
                file_name = bm.split(".")[0]
                if check_if_file_is_in_source_code(file_name, project, buggy_commit):
                    coverage_sum += float(coverages[file_name + ".java"])

            coverage_average = coverage_sum/len(buggy_methods)
            print(coverage_average)

def print_average_coverage_of_all_methods():
    averages = {}
    for bug_report_analysis_file in glob.glob(ochiai_path + "*/*.json"):
        project = bug_report_analysis_file.split("/")[-2]
        bug_id = bug_report_analysis_file.split("/")[-1].replace(".json","")

        print("---------------")
        print(project + " - " + bug_id)
        buggy_commit = t_evos_bugs_information[project][bug_id]["commit_with_bug"]
        coverage_details_folder = coverage_data_folder + project + "/" + buggy_commit + "/details/"
        if project not in t_evos_bugs_information.keys():
            continue
        if bug_id not in t_evos_bugs_information[project].keys():
            continue
        ochiai_scores_data = json_file_to_dict(bug_report_analysis_file)
        coverage_sum = 0
        print(len(ochiai_scores_data.keys()))
        for method_key in ochiai_scores_data.keys():
            print(method_key)
            file_name = method_key.split(".")[0]
            method_name = method_key.split(".")[1]
            coverage_sum += get_method_coverage(file_name, method_name, None, coverage_details_folder, project, buggy_commit)

        coverage_average = coverage_sum/len(ochiai_scores_data.keys())
        print("---------------")
        print(project + " - " + bug_id)
        print("!!!!!!" + str(coverage_average))
        averages[project + "-" + bug_id] = coverage_average
    dict_to_json_file("all_methods_average", averages, base_path + "BugReportsMining/output/")


def print_average_coverage_of_all_files():
    for project in t_evos_bugs_information.keys():
        for bug_id in t_evos_bugs_information[project].keys():
            print("---------------")
            print(project + " - " + bug_id)
            buggy_commit = t_evos_bugs_information[project][bug_id]["commit_with_bug"]
            csv_file_path = coverage_folder + project+ "/" + buggy_commit +'.csv'
            if os.path.exists(csv_file_path):
                coverages = {}
                with open(os.path.join(csv_file_path), 'r') as file:
                    csv_reader = csv.reader(file)
                    count = 0
                    for row in csv_reader:
                        if count > 0:
                            coverages[row[0]] = row[1]
                        count = count + 1
            coverage_sum = 0
            for file in coverages.keys():
                coverage_sum += float(coverages[file])

            coverage_average = coverage_sum/len(coverages)
            print(coverage_average)

#print("print_best_buggy_method_positions_for_each_Ochiai")
#print_best_buggy_method_positions_for_each_Ochiai()
#print("-------")
#print("print_average_coverage_of_the_methods_in_the_stack_trace")
#print_average_coverage_of_the_methods_in_the_stack_trace()
#print("-------")
#print("print_average_coverage_of_the_buggy_methods")
#print_average_coverage_of_the_buggy_methods()
#print("-------")
#print("print_average_coverage_of_all_methods")
#print_average_coverage_of_all_methods()
#print("-------")
#print("print_average_coverage_of_the_files_in_the_stack_trace")
#print_average_coverage_of_the_files_in_the_stack_trace()
#print("-------")
#print("print_average_coverage_of_the_buggy_files")
#print_average_coverage_of_the_buggy_files()
print("-------")
print("print_average_coverage_of_all_files")
print_average_coverage_of_all_files()
```
